# **3) Feature Engineering in the Data :-**
Adding the following features :- *Feature Name *Utilization_Ratio *Delinquency_Count *Longest_Delinquency_Streak *Repayment_Consistency *Overpayment_Frequency *Repay_Ratio_1 to 6

```python
# Count of late payments (status > 0 means late)

pay_cols = ['pay_0' , 'pay_2' , 'pay_3' , 'pay_4' , 'pay_5' , 'pay_6']
train_df['total_late_payments'] = (train_df[pay_cols] > 0).sum(axis=1)

# Max delinquency severity
train_df['max_delinquency'] = train_df[pay_cols].max(axis=1)

# Current delinquency streak
def get_current_streak(row):
    streak = 0
    for i in pay_cols:  # Check PAY_0 first (most recent)
        if row[f'{i}'] > 0:
            streak += 1
        else:
            break
    return streak

train_df['current_delinquency_streak'] = train_df.apply(get_current_streak, axis=1)
train_df
```

```python
# Count of months with only minimum payment (status = 0)
pay_cols = ['pay_0' , 'pay_2' , 'pay_3' , 'pay_4' , 'pay_5' , 'pay_6']
train_df['min_payment_months'] = (train_df[pay_cols] == 0).sum(axis=1)

# Count of months with full payment (status < 0)
train_df['full_payment_months'] = (train_df[pay_cols] < 0).sum(axis=1)

# Count of months with delayed payment (status > 0)
train_df['delayed_months'] = (train_df[pay_cols] > 0).sum(axis=1)

# Payment consistency
pay_amt_cols = [f'pay_amt{i}' for i in range(1, 7)]
train_df['payment_consistency'] = train_df[pay_amt_cols].std(axis=1)
train_df
```

```python
import numpy as np
import pandas as pd
from sklearn.impute import SimpleImputer

# --- Step 1: Define Feature Engineering Function ---
def feature_engineering(df):
    df = df.copy()

    # Utilization Ratio
    df['Utilization_Ratio'] = df['AVG_Bill_amt'] / df['LIMIT_BAL'].replace(0, np.nan)

    # Delinquency Count
    pay_cols = ['pay_0', 'pay_2', 'pay_3', 'pay_4', 'pay_5', 'pay_6']
    df['Delinquency_Count'] = df[pay_cols].ge(1).sum(axis=1)

    # Longest Delinquency Streak
    df['Longest_Delinquency_Streak'] = df[pay_cols].max(axis=1)

    # Repayment Consistency
    pay_amt_cols = [f'pay_amt{i}' for i in range(1, 7)]
    df['Repayment_Consistency'] = df[pay_amt_cols].std(axis=1)

    # Overpayment Frequency
    bill_amt_cols = [f'Bill_amt{i}' for i in range(1, 7)]
    df['Overpayment_Frequency'] = df[bill_amt_cols].lt(0).sum(axis=1)

    # Monthly Repayment Ratios
    for i in range(1, 7):
        bill_col = f'Bill_amt{i}'
        pay_col = f'pay_amt{i}'
        ratio_col = f'Repay_Ratio_{i}'
        df[ratio_col] = df[pay_col] / df[bill_col].replace(0, np.nan)

    return df

#  Apply Feature Engineering to both sets ---
train_df = feature_engineering(train_df)
validation_df = feature_engineering(validation_df)
```

# **4) Data Pre-Processing**



```python
train_df_clean = train_df.drop(columns='Customer_ID', errors='ignore').copy()
validation_df_clean = validation_df.drop(columns='Customer_ID', errors='ignore').copy()


#train_data_clean = train_data.drop(columns='Customer_ID', errors='ignore').copy()
#validation_data_clean = validation_data.drop(columns='Customer_ID', errors='ignore').copy()

# filling missing data from our coloumn
from sklearn.impute import SimpleImputer

numeric_cols = [
    'LIMIT_BAL', 'age', 'AVG_Bill_amt', 'PAY_TO_BILL_ratio',
    'Utilization_Ratio', 'Delinquency_Count', 'Longest_Delinquency_Streak',
    'Repayment_Consistency', 'Overpayment_Frequency'
] + [f'pay_amt{i}' for i in range(1, 7)] + [f'Bill_amt{i}' for i in range(1, 7)] + \
    [f'Repay_Ratio_{i}' for i in range(1, 7)]

imputer = SimpleImputer(strategy='mean')

train_df_clean[numeric_cols] = imputer.fit_transform(train_df_clean[numeric_cols])
validation_df_clean[numeric_cols] = imputer.transform(validation_df_clean[numeric_cols]) # Use transform only for validation
```

```python
# Define categorical columns to encode
categorical_cols = ['sex', 'education', 'marriage']

#Apply one-hot encoding to both train and validation datasets
train_data_encoded = pd.get_dummies(train_df_clean.copy(), columns=categorical_cols, drop_first=True)
validation_data_encoded = pd.get_dummies(validation_df_clean.copy(), columns=categorical_cols, drop_first=True)

# Align the columns in validation with train
train_data_encoded, validation_data_encoded = train_data_encoded.align(validation_data_encoded, join='left', axis=1, fill_value=0)

print("Train shape:", train_data_encoded.shape)
print("Validation shape:", validation_data_encoded.shape)
```

```python
# Separate X and y *before* handling missing values with the imputer
X_train = train_data_encoded.drop(columns='next_month_default')
y_train = train_data_encoded['next_month_default']

# Initialize and fit the imputer on the training features (X_train)
imputer = SimpleImputer(strategy='mean')
X_train_imputed = imputer.fit_transform(X_train)

# Convert the imputed array back to a DataFrame to maintain column names
X_train_imputed = pd.DataFrame(X_train_imputed, columns=X_train.columns)

# Initialize SMOTE
smote = SMOTE(random_state=42)

# Fit and resample using the imputed data
X_train_resampled, y_train_resampled = smote.fit_resample(X_train_imputed, y_train)
```

## **5) Model Building**

```python
model_results = {}
from sklearn.metrics import fbeta_score, precision_recall_curve, average_precision_score

from sklearn.model_selection import cross_val_score, cross_val_predict
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score, f1_score, roc_auc_score,
    confusion_matrix, roc_curve
)
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import time

def eval_model(model, X_train, y_train, model_name):
    print(f"\n Evaluating: {model_name}")
    start = time.time()

    # Cross-validated AUC (uses predict_proba internally)
    auc_scores = cross_val_score(model, X_train, y_train, cv=3, scoring='roc_auc')

    # Cross-validated Predictions (hard labels)
    y_pred = cross_val_predict(model, X_train, y_train, cv=3)

    # Cross-validated Probabilities (for ROC curve)
    y_proba = cross_val_predict(model, X_train, y_train, cv=3, method='predict_proba')[:, 1]

    # Metrics
    acc = accuracy_score(y_train, y_pred)
    prec = precision_score(y_train, y_pred, zero_division=0)
    rec = recall_score(y_train, y_pred, zero_division=0)
    f1 = f1_score(y_train, y_pred, zero_division=0)
    auc = roc_auc_score(y_train, y_proba)

    end = time.time()

    model_results[model_name] = {
        'model': model,
        'X': X_train,
        'y_true': y_train,
        'y_pred': y_pred,
        'y_proba': y_proba,
        'auc': auc
    }

    # Output
    print(f"AUC Scores (5-fold): {auc_scores}")
    print(f"Mean AUC: {auc_scores.mean():.4f}")
    print(f"Accuracy: {acc:.4f}")
    print(f"Precision: {prec:.4f}")
    print(f"Recall: {rec:.4f}")
    print(f"F1 Score: {f1:.4f}")
    print(f"Time: {end - start:.2f} sec")

    # Confusion Matrix
    cm = confusion_matrix(y_train, y_pred)
    plt.figure(figsize=(5, 4))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=[0, 1], yticklabels=[0, 1])
    plt.title(f'{model_name} - Confusion Matrix')
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.show()

    # ROC Curve
    fpr, tpr, thresholds = roc_curve(y_train, y_proba)
    plt.figure(figsize=(6, 5))
    plt.plot(fpr, tpr, label=f'AUC = {auc:.2f}')
    plt.plot([0, 1], [0, 1], 'k--')  # Diagonal
    plt.title(f'{model_name} - ROC Curve')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
from lightgbm import LGBMClassifier
# 1. Logistic Regression
log_reg = LogisticRegression(max_iter=1000, class_weight='balanced')
# Use the resampled data for cross-validation
eval_model(log_reg, X_train_resampled, y_train_resampled, "Logistic Regression")

# 2. Decision Tree
tree = DecisionTreeClassifier(class_weight='balanced', random_state=42)
# Use the resampled data for cross-validation
eval_model(tree, X_train_resampled, y_train_resampled, "Decision Tree")

# 3. Random Forest
forest = RandomForestClassifier(n_estimators=100, class_weight='balanced', random_state=42)
# Use the resampled data for cross-validation
eval_model(forest, X_train_resampled, y_train_resampled, "Random Forest")

# 4. XGBoost
xgb = XGBClassifier(n_estimators=100, eval_metric='logloss', use_label_encoder=False)
eval_model(xgb, X_train_resampled, y_train_resampled, "XGBoost")
log_reg = LogisticRegression(max_iter=1000, class_weight='balanced')
eval_model(log_reg, X_train, y_train, "Logistic Regression")

# 5. LightGBM
lgb = LGBMClassifier(n_estimators=100, class_weight='balanced', random_state=42)
eval_model(lgb, X_train_resampled, y_train_resampled, "LightGBM")
```

## **6) Model Evaluation**

```python
# Find the model with highest AUC
best_model_name = max(model_results, key=lambda name: model_results[name]['auc'])
best_model_info = model_results[best_model_name]

print(f"\n Best Model Based on AUC: {best_model_name} (AUC = {best_model_info['auc']:.4f})")

# Optional: Calculate F2-score for each model to emphasize recall
from sklearn.metrics import fbeta_score

print("\n F2 Scores (emphasizing Recall):")
for model_name, metrics in model_results.items():
    f2 = fbeta_score(metrics['y_true'], metrics['y_pred'], beta=2)
    print(f"{model_name}: F2 Score = {f2:.4f}")

# Precision-Recall Curve for Best Model
from sklearn.metrics import precision_recall_curve, average_precision_score
import matplotlib.pyplot as plt
import numpy as np

best_model = best_model_info['model']
X = best_model_info['X']
y_true = best_model_info['y_true']

# --- Fit the best model explicitly ---
print(f"\n Fitting the best model ({best_model_name}) before plotting PR curve...")
best_model.fit(X, y_true)

# Get probability scores for PR curve
if hasattr(best_model, "predict_proba"):
    y_scores = best_model.predict_proba(X)[:, 1]
elif hasattr(best_model, "decision_function"):
    y_scores = best_model.decision_function(X)
else:
    raise ValueError("Model doesn't support probability scoring.")

precision, recall, thresholds = precision_recall_curve(y_true, y_scores)
avg_prec = average_precision_score(y_true, y_scores)

# Plot the PR curve
plt.figure(figsize=(10, 6))
plt.plot(recall, precision, label=f'PR Curve (AP = {avg_prec:.4f})')
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title(f'Precision-Recall Curve - {best_model_name}')
plt.legend()
plt.grid(True)
plt.show()

# --- Threshold Analysis ---
# Find the threshold with highest F2 score
f2_scores = [fbeta_score(y_true, y_scores >= t, beta=2, zero_division=0) for t in thresholds]
best_threshold_idx = np.argmax(f2_scores)
best_threshold = thresholds[best_threshold_idx]

print(f"\n Best Threshold for F2 Score: {best_threshold:.4f} (F2 = {f2_scores[best_threshold_idx]:.4f})")

# Plot F2 score vs. threshold
plt.figure(figsize=(10, 5))
plt.plot(thresholds, f2_scores, marker='o')
plt.axvline(best_threshold, color='red', linestyle='--', label=f'Best Threshold = {best_threshold:.4f}')
plt.title(f'F2 Score vs Threshold - {best_model_name}')
plt.xlabel('Threshold')
plt.ylabel('F2 Score')
plt.legend()
plt.grid(True)
plt.show()
```
